```python
import numpy as np
import matplotlib.pyplot as plt

T = 7.24e-6
K = 800e9
Fs = K*T*5
N = int((Fs)*T)
t = np.linspace(-T/2, T/2, N)

T_rect = T;
rect = abs(t) < T_rect/2;

s = np.exp(1j*(np.pi*K*(t)**2))*rect
s1=np.exp(1j*(np.pi*K*(t)**2))

fig, axs = plt.subplots(3, 1, figsize=(8,8))

axs[0].plot(t, s.real)
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Real part')
axs[0].set_title('Real part of Linear FM signal')

axs[1].plot(t, s.imag)
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('Imaginary part')
axs[1].set_title('Imaginary part of Linear FM signal')

axs[2].plot(t, np.abs(s))
axs[2].set_xlabel('Time (s)')
axs[2].set_ylabel('Magnitude')
axs[2].set_title('Magnitude of Linear FM signal')

plt.tight_layout()
plt.show()
```

```python
beta_values = [1, 2.5, 5]
oversampling_factor = 5

for beta in beta_values:
    T = 7.24e-6
    K = 800e9
    Fs = K*T*5
    N = int((Fs)*T)
    t = np.linspace(-T/2, T/2, N)

    T_rect = T;
    rect = abs(t) < T_rect/2;

    s = np.exp(1j*(np.pi*K*(t)**2))*rect

    # apply Kaiser window
    w = np.kaiser(N, beta)
    s = s * w

    fig, axs = plt.subplots(3, 1, figsize=(8,9))

    axs[0].plot(t, s.real)
    axs[0].set_xlabel('Time (s)')
    axs[0].set_ylabel('Real part')
    axs[0].set_title(f'Real part of Linear FM signal with Kaiser window (beta={beta})')

    s_fft = np.fft.fftshift(np.fft.fft(s))

    freq = np.linspace(-Fs/2, Fs/2, N)

    axs[1].plot(freq, s_fft.imag)
    axs[1].set_xlabel('Frequency (Hz)')
    axs[1].set_ylabel('Imaginary part')
    axs[1].set_title(f'Imaginary part of Linear FM signal with Kaiser window (beta={beta})')


    axs[2].plot(freq, np.abs(s_fft))
    axs[2].set_xlabel('Frequency (Hz)')
    axs[2].set_ylabel('Magnitude')
    axs[2].set_title(f'Magnitude of Linear FM signal with Kaiser window (beta={beta})')


    plt.tight_layout()
    plt.show()
```
